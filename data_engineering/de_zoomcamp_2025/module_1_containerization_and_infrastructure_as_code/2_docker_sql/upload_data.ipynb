```python
df_green = pd.read_csv('green_tripdata_2019-10.csv.gz')
df_green.head()
```

```python
df_zones = pd.read_csv('taxi_zone_lookup.csv')
df_zones.head()
```

```python
def segmentation_count(value):
    if value <= 1:
        return 1
    elif value <=3:
        return 2
    elif value <=7:
        return 3
    elif value <=10:
        return 4
    else:
        return 5
        
```

```python
df_green['segmentation'] = df_green['trip_distance'].apply(lambda x:segmentation_count(x))
```

```python
df_green[(df_green['lpep_pickup_datetime'] >= '2019-10-01') & (df_green['lpep_dropoff_datetime'] < '2019-11-01')].groupby('segmentation').count()
```
